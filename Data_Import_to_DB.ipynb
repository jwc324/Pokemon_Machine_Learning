```python
#import libraries
import pandas as pd #to read files
from sqlalchemy import create_engine #to import into database
from config import pokedb_password
```

```python
#import data from csv files
poketypes_df = pd.read_csv("Database/data_files/pokedex_types.csv")
pokeagainst_df = pd.read_csv("Database/data_files/pokedex_against.csv")
```

```python
#confirm successful creation
poketypes_df.head()
```

```python
pokeagainst_df.head()
```

```python
#Join poketypes_df and pokeagainst_df together into new df & confirm successful creation
poke_df = pd.merge(poketypes_df, pokeagainst_df)
poke_df.dtypes
```

```python
##import modules into database
db_string = f"postgresql://postgres:{pokedb_password}@localhost:5432/pokemon_421_db"
```

```python
#create database engine
engine = create_engine(db_string)
```

```python
#import new poke_df to database
poke_df.to_sql(name="poke_df", con=engine)
```
